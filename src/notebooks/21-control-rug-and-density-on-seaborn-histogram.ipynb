## Histogram + rug + kernel density curve

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")
df = sns.load_dataset("iris")

sns.displot(data=df["sepal_length"], kde=True, rug=True)
plt.show()
```

## Customizing rugs

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set_theme(style="darkgrid")
df = sns.load_dataset("iris")

sns.displot(df["sepal_length"],
             kde=True,
             rug=True,
             rug_kws={"color": "r", "alpha": 0.8, "linewidth": 2, "height":0.2})
plt.show()
```

## Customizing the density curve

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set_theme(style="darkgrid")
df = sns.load_dataset("iris")

sns.distplot(df["sepal_length"],
             kde=True,
             kde_kws={"color": "g", "alpha": 0.3, "linewidth": 5, "fill": True})
plt.show()
```
