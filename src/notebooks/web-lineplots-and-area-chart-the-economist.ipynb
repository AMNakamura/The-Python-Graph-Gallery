```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import lines
from matplotlib import patches
from matplotlib.patheffects import withStroke
```

```python
year = [2008, 2012, 2016, 2020]
latin_america = [10, 9, 7.5, 5.8]
asia_and_pacific = [13.5, 9.5, 7.5, 5.5]
sub_saharan_africa = [25.5, 21, 22.2, 24]
percentages = [latin_america, asia_and_pacific, sub_saharan_africa]
```

```python
BROWN = "#AD8C97"
GREEN = "#2FC1D3"
BLUE = "#076FA1"
colors = [BROWN, GREEN, BLUE]
```

```python
fig, ax = plt.subplots(figsize=(8, 6))


for y, color in zip(percentages, colors):
    ax.plot(year, y, color=color, lw=5)
    ax.scatter(year, y, fc=color, s=100, lw=1.5, ec="white", zorder=12)


ax.yaxis.set_ticks([i * 5 for i in range(0, 7)])
ax.yaxis.set_ticklabels([i * 5 for i in range(0, 7)])
ax.yaxis.set_tick_params(labelleft=False, length=0)

ax.xaxis.set_ticks([2008, 2012, 2016, 2020])
ax.xaxis.set_ticklabels([2008, 12, 16, 20], fontsize=16, fontfamily="Econ Sans Cnd", fontweight=100)
ax.xaxis.set_tick_params(length=6, width=1.2)


# Set whether axis ticks and gridlines are above or below most artists.
ax.set_axisbelow(True)
ax.grid(axis = "y", color="#A8BAC4", lw=1.2)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_lw(1.2)
ax.spines["bottom"].set_capstyle("butt")

ax.set_ylim(0, 35)
ax.set_xlim(2007, 2021.5)

PAD = 0.4
for label in [i * 5 for i in range(0, 7)]:
    ax.text(
        2021.5, label + PAD, label, 
        ha="right", va="baseline", fontsize=16,
        fontfamily="Econ Sans Cnd", fontweight=100
    )


# Add annotations for regions    
ax.text(
    2007.9, 6, "Latin America and\nthe Caribbean", 
    color=BROWN, fontfamily="Econ Sans Cnd", 
    fontsize=18, va="center",
    path_effects=[withStroke(linewidth=12, foreground="white")]
) 

ax.text(
    2007.9, 26.5, "Sub-Saharan Africa", 
    color=BLUE, fontfamily="Econ Sans Cnd", 
    fontsize=18, va="baseline"
) 

ax.text(
    2010, 12.5, "Asia and the Pacific", 
    color=GREEN, fontfamily="Econ Sans Cnd", 
    fontsize=18, va="baseline"
)
 
# Add other annotations outside the plot region
ax.text(0, 0.975, "Selected regions,", 
        fontsize=18, fontfamily="Econ Sans Cnd", va="top",
        transform=ax.transAxes, fontweight="bold" 
)
ax.text(0.3, 0.975, "% of child population", 
        fontsize=18, fontfamily="Econ Sans Cnd", va="top",
        transform=ax.transAxes
)

ax.add_artist(
    lines.Line2D(
        [0, 0.05], [1, 1], lw=2, color="black",
        solid_capstyle="butt", transform=ax.transAxes
    )
)

fig.set_facecolor("white")
fig.savefig("plot.png", dpi=300)
```
