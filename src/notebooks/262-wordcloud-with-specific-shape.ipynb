## Load an image

First, we need to choose an image. Ideally, it's in **black and white**. In our case, we will use a wave image. Make sure you have the image **locally stored** on your computer.


```python
import numpy as np
from PIL import Image

wave_mask = np.array(Image.open( "wave.jpg"))
```

## Simple wordcloud

Let's see what a basic [wordcloud](https://python-graph-gallery.com/wordcloud/) looks like. We will use the `wordcloud` library to generate the wordcloud:


```python
# Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt
 
text=("""Python Python Python Matplotlib Matplotlib Seaborn Network Plot Violin Chart Pandas Datascience Wordcloud Spider Radar Parrallel Alpha Color Brewer Density Scatter Barplot Barplot Boxplot Violinplot Treemap Stacked Area Chart Chart Visualization Dataviz Donut Pie Time-Series Wordcloud Wordcloud Sankey Bubble""")
 
# Create the wordcloud object
wordcloud = WordCloud(
    mask=None,
    background_color='white'
).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## Add a mask

For using a shape for the wordcloud, we need to use a `mask`, which is a black and white image. We will use the Twitter logo as a mask.


```python
# Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text=("""Python Python Python Matplotlib Matplotlib Seaborn Network Plot Violin Chart Pandas Datascience Wordcloud Spider Radar Parrallel Alpha Color Brewer Density Scatter Barplot Barplot Boxplot Violinplot Treemap Stacked Area Chart Chart Visualization Dataviz Donut Pie Time-Series Wordcloud Wordcloud Sankey Bubble""")

# Create the wordcloud object
wordcloud = WordCloud(
    mask=wave_mask,
    background_color='white'
).generate(text)

# Display the generated image:
fig, ax = plt.subplots(1, 1, figsize=(6, 8))
ax.imshow(wordcloud, interpolation="bilinear")
ax.axis("off")
plt.show()
```
